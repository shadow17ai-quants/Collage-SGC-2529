```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("PNRao_Education_Fee_Payments.csv")
```

```python
df.isnull().sum()
```

```python
df['PaymentDate'] = pd.to_datetime(df['PaymentDate'])
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Method'].value_counts()
```

```python
df['Semester'].value_counts()
```

```python
df['Month'] = df['PaymentDate'].dt.to_period('M')
```

```python
df.groupby('Semester')['Amount'].sum()
```

```python
df.groupby('Method')['Amount'].sum()
```

```python
monthly_trend = df.groupby('Method')['Amount'].sum()
```

```python
monthly_trend.plot(kind='line', marker='o' , figsize =(10,5), title='Monthly Revenue Trend')
```

```python
df['Method'].value_counts().plot(kind='pie', autopct='1.1f%%' , figsize= (6, 6) , title= ' Payment Method Distribution')
```

```python
df.groupby('Semester') ['Amount'].plot(kind='bar',title= 'Total Revenue By Semester',color = ['red','blue'])


```
